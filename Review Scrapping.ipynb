```python
date = soup.find_all('span', class_='ca-txt-cpt')
print(date)
```

```python
review_date =[]
for i in date:
    date_string = (i.text.split(':'))[1]
    review_date.append(date_string)
review_date
```

```python
strong = soup.find_all('strong',class_="rvw-aut__inf-nm")
name=[]
reviewer_location=[]
for i in strong:
    x = ((i.text.split('of')))
    name.append(x[0])
    reviewer_location.append(x[1])
    
print(name)
print(reviewer_location)
```

```python
reviews=[]
p_tags = soup.find_all('div', class_= 'rvw-bd')
for tag in p_tags:
    x = (tag.text.split('2020'))
    reviews.append(x[1])

reviews
```
